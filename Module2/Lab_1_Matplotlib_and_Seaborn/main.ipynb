#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

**Hints**:

```python
# MEDIAN STEPS BY MONTH_ENCODED
# aggregate median
fitbit_month_median = fitbit[["Steps","Months_encoded"]].groupby(["Months_encoded"]).agg("median")
fitbit_workweek_median = fitbit[["Steps","Work_or_Weekend"]].groupby(["Work_or_Weekend"]).agg("median")

#creation plots
fitbit_month_median.plot(figsize=(12,4))
fitbit_workweek_median.plot.bar(figsize=(12,4))
plt.subplots(1,2)
    
plt.show()
# save figures
plt.savefig('figures2.png')
```
