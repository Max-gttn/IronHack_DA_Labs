```python
import pandas as pd
```

```python
df=pd.DataFrame(
    [
    (1, 231, 'self-employed','009', 'high', 2, 30200),
    (2, 765, 'students', '005', 'high', 3, 12700),
    (3, 453, 'Horeca', '007', 'medium', 5, 89400),
    (4, 231, 'self-employed', '009', 'high', 2, 30200),
    (5, 892, 'finance', '003', 'low', 3, 740000)
    ],
    columns=('TransactionID', 'ClientID', 'Profession', 'Bank_dep', 'Risk', 'Number of credits', 'Revenue')
    )
df
```

```python
#Dummy Encoder
dummy = pd.get_dummies(df)
dummy
```

```python
#Label Encoder
from sklearn.preprocessing import LabelEncoder

df_copy=df.copy()
label= LabelEncoder()
df_copy['Risk_encode']=label.fit_transform(df_copy['Risk'])
df_copy
```

```python
# Backward Difference Coding 
import category_encoders as ce

cat_df = df.select_dtypes(include=['object']).copy()
encoder = ce.BackwardDifferenceEncoder(cols=['Profession'])
df_bd = encoder.fit_transform(cat_df)

df_bd.head()
```

```python
#BaseN
import category_encoders as ce
# create an object of the BaseNEncoder
ce_baseN4 = ce.BaseNEncoder(cols=['Profession'],base=4)
# fit and transform and you will get the encoded data
df_BaseN = ce_baseN4.fit_transform(df)
df_BaseN
```

```python
#Binary
# create object of BinaryEncoder
ce_binary = ce.BinaryEncoder(cols = ['Profession','Risk'])
df_binary = ce_binary.fit_transform(df)
df_binary
```
