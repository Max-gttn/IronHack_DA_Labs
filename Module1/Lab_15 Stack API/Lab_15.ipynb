Question 1: Find the questions and answers of last month.

```python
SITE = StackAPI('stackoverflow')
questions = SITE.fetch('questions', fromdate=1636239600, todate=1638831600,sort='votes')

df = pd.DataFrame(questions)
df
```

```python
def flatten(df, col_list):
    for column in col_list:
        flattened = pd.DataFrame(dict(df[column])).transpose()
        columns = [str(col) for col in flattened.columns]
        flattened.columns = [column + '_' + colname for colname in columns]
        data = pd.concat([df, flattened], axis=1)
        data = data.drop(column, axis=1)
    return data
nested_columns = ['items']

flat = flatten(df, nested_columns)


flat.head(10)
```
